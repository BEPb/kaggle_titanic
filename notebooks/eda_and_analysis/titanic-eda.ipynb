## I keep getting Age's missing value error.
* Let's compare the values after calibration.

```python
# Fill in the missing age data with the median value.
train['Age_median'] = train['Age']
train['Age_mean'] = train['Age']
train['Age_mode'] = train['Age']

train['Age_median'].fillna(train['Age'].median(), inplace = True)
train['Age_mean'].fillna(train['Age'].mean(), inplace = True)
train['Age_mode'].fillna(train['Age'].mode()[0], inplace = True)
```

```python
(ggplot(train) 
 + aes(x='Age')
 + geom_histogram(binwidth=10)
 + ggtitle('age distribution')
 + theme(text=element_text(family='NanumBarunGothic'))
)
```

```python
(ggplot(train) 
 + aes(x='Age_median')
 + geom_histogram(binwidth=10)
 + ggtitle('Median correction for age group missing values')
 + theme(text=element_text(family='NanumBarunGothic'))
)
```

```python
(ggplot(train) 
 + aes(x='Age_mean')
 + geom_histogram(binwidth=10)
 + ggtitle('Corrected for age group missing values with mean value')
 + theme(text=element_text(family='NanumBarunGothic'))
)
```

```python
(ggplot(train) 
 + aes(x='Age_mode')
 + geom_histogram(binwidth=10)
 + ggtitle('Age group missing value corrected for the mode')
 + theme(text=element_text(family='NanumBarunGothic'))
)
```
